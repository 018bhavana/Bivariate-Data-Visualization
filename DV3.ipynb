```python
import pandas as pd
# Load the dataset
df = pd.read_csv("/content/student-mat (1).csv")
print(df.head())
```

```python
# Categorical = object/string columns
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
# Continuous = int or float columns
continuous_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
print("Categorical Columns:", categorical_cols)
print("Continuous Columns:", continuous_cols)
```

```python
import matplotlib.pyplot as plt
pd.crosstab(df['sex'], df['school']).plot(kind='bar', stacked=True, color=['skyblue','orange'])
plt.title("Stacked Bar Chart: Gender vs Approved")
plt.ylabel("Count")
plt.show()
```

```python
pd.crosstab(df['sex'], df['school']).plot(kind='bar',stacked=False, color=['skyblue','orange'])
plt.title("Grouped Bar Chart: Gender vs Approved")
plt.ylabel("Count")
plt.show()
```

```python
(pd.crosstab(df['sex'], df['school'], normalize='index') * 100).plot(kind='bar',stacked=True)
plt.title("Segmented Bar Chart (%): Gender vs Approved")
plt.ylabel("Percentage")
plt.show()
```

```python
import seaborn as sns
sns.lmplot(data=df, x='absences', y='G3', line_kws={'color':'red'})
plt.title("Scatterplot with Fit Line: Absences vs Final Grade")
plt.show()
```

```python
df.groupby('sex')['G3'].mean().plot(kind='bar', color='lightgreen')
plt.title("Average Final Grade by Gender")
plt.ylabel("Average Grade")
plt.show()
```

```python
sns.kdeplot(data=df, x='G3', hue='sex', fill=True)
plt.title("Grouped Kernel Density: Grade by Gender")
plt.show()
```

```python
sns.boxplot(data=df, x='sex', y='G3')
plt.title("Box Plot: Gender vs Final Grade")
plt.show()
```

```python
sns.violinplot(data=df, x='sex', y='G3', inner='quartile')
plt.title("Violin Plot: Gender vs Final Grade")
plt.show()
```
